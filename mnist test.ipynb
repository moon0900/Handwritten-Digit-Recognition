```python
import os, shutil
import matplotlib.pyplot as plt
import numpy as np

from keras.preprocessing.image import ImageDataGenerator
```

```python
# 원본 데이터셋 디렉터리 경로
original_dataset_dir = './mydataset'
os.getcwd()
```

```python
os.chdir('/content/drive/MyDrive/Colab Notebooks/2022-DL/')
```

```python
def invert(image):
    new_img = 255 - image
    return new_img

test_datagen = ImageDataGenerator(rescale=1./255, preprocessing_function=invert)

test_generator = test_datagen.flow_from_directory(
        original_dataset_dir,
        target_size=(28, 28),
        color_mode='grayscale',
        batch_size=5,
        class_mode='categorical')
```

```python
from tensorflow.python.keras.models import load_model
model = load_model('mnist_convnet.h5')
```

```python
test_generator.reset()

print("-- Evaluate --")
scores = model.evaluate(test_generator, steps=60)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))

print("-- Predict --")

wrong_predict = []
for i in range(10):
  wrong_predict.append([])

```

```python
for j in range(60):
  x_val,y_val = test_generator.next()
  y_pred = model.predict(x_val)
  for i in range(len(y_pred)):
    predict = np.argmax(y_pred[i])
    y_true = np.argmax(y_val[i])
    if predict != y_true:
      wrong_predict[y_true].append(predict)
```

```python
for idx, pred in enumerate(wrong_predict):
  print(idx,'에 대한 오답 수:', len(pred))
  if pred:
    print(' - 오답들: ', pred)
  print()
```

```python
import os, shutil, glob

# 분할 데이터셋을 저장할 디렉터리
base_dir = '/content/drive/Shareddrives/Deep Learning/2022-DL/OurDataset/'
train_dir = os.path.join(base_dir, 'train/')
test_dir = os.path.join(base_dir, 'test/')
dataadd = '/content/drive/MyDrive/mynet/datasets/new/'
```

```python
dataset_source = '/content/drive/Shareddrives/Deep Learning/2022-DL/'
```

```python
for d in ['2차', 'JDATASET2', 'mydataset', 'Jdataset', '손글씨데이터', '3차']:
  dir = os.path.join(dataset_source, d)
  for i in range(10):
      fnames = os.listdir(dir+'/'+str(i)+"/")
      for fname in fnames:
        src = os.path.join(dir+'/'+str(i), fname)
        dst = os.path.join(train_dir+str(i), d+'_'+fname)
        shutil.copyfile(src, dst)
```

```python
# 각 데이터셋에서 train_dir로 복사
for d in ['train','validation','test']:
  for i in range(10):
      dir = os.path.join(dataadd, d)
      dir = os.path.join(dir, str(i))
      fnames = os.listdir(dir)
      for fname in fnames:
        src = os.path.join(dir, fname)
        dst = os.path.join(base_dir, 'train')
        dst = os.path.join(dst, str(i))
        dst = os.path.join(dst, d[:3]+'_new_'+fname)
        print(src, dst)
        shutil.copyfile(src, dst)
```

```python

if os.path.exists(base_dir):
    shutil.rmtree(base_dir)
os.mkdir(base_dir)

# 훈련, 검증, 테스트 분할을 위한 디렉터리
train_dir = os.path.join(base_dir, 'train/')
os.mkdir(train_dir)
test_dir = os.path.join(base_dir, 'test/')
os.mkdir(test_dir)

# 레이블 디렉터리 생성
for i in range(10):
    train_label_dir = os.path.join(train_dir, str(i))
    test_label_dir = os.path.join(test_dir, str(i))
    os.mkdir(train_label_dir)
    os.mkdir(test_label_dir)

```

```python
print('레이블 별 데이터 수')
for i in range(10):
  fnames = os.listdir(train_dir+str(i)+"/")
  print(i, len(fnames))
```

```python
import random

for i in range(10):
  fnames = os.listdir(train_dir+str(i)+"/")
  #print(fnames)
  for f in random.sample(fnames, 20):

    src = os.path.join(train_dir+str(i), f)
    dst = os.path.join(test_dir+str(i), f)
    print(src, dst)
    shutil.move(src, dst)
```
